```python
import glob
import pandas as pd
import numpy as np
```

```python
Region = "Massachusetts"
State = "MA"
```

## Prepare Flight Files

```python
# Import Flights Files

root = r'c:\\users\\user\\Documents\\JCU\\MA 5853 Data Science Project 1\\Data\\Airport analysis\\'
path = r'\\Uploads2\\*Flights*.csv'

files = glob.glob(root+Region+path)
dfs = [pd.read_csv(f, sep=",") for f in files]

df = pd.concat(dfs,ignore_index=True)

df = df.drop(df.columns[[0]], axis=1)

df = df.rename({'Flights': 'Actual','Forecast':'ForecastO','Category':'Segment'}, axis=1)

df['Category'] = "Flights"

df['Origin'] = State

df.date = pd.to_datetime(df.ds)

df['Qrt'] = pd.PeriodIndex(df.date, freq='Q')
df['Qrt'] =  df['Qrt'].astype(str)
df['Qrt'] = df['Qrt'].str.replace(r'(?<=\d)(?=\D)',' ', regex=True)

df['Forecast'] = np.where(df['Actual'].isnull(), df['ForecastO'], np.nan )      

df['Datestamp'] = df['ds'] 

print(df.shape)
```

```python
df =df.loc[:,["Datestamp","Airline","Segment","Actual","ForecastO","YOYDiff","Forecast","Qrt","ds","Category","Origin"]]
dfFlights = df
dfFlights
```
