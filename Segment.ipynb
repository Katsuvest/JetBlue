```python
import glob
import pandas as pd

catchment = "\\Massachusetts\\"

root = r'c:\\users\\user\\Documents\\JCU\\MA 5853 Data Science Project 1\\Data\\Airport analysis\\'
path = r'\\Segment T100\\*.csv'

files = glob.glob(root+catchment+path)
dfs = [pd.read_csv(f, sep=",") for f in files]

df = pd.concat(dfs,ignore_index=True)


print(df.shape)
```

## Filter for those airlines needed for review

```python
df2 = df


AA = df[df['UNIQUE_CARRIER']=='AA']
AS = df[df['UNIQUE_CARRIER']=='AS']
B6 = df[df['UNIQUE_CARRIER']=='B6']
DL = df[df['UNIQUE_CARRIER']=='DL']
NK = df[df['UNIQUE_CARRIER']=='NK']
UA = df[df['UNIQUE_CARRIER']=='UA']
WN = df[df['UNIQUE_CARRIER']=='WN']

df = pd.concat([AA, AS, B6, DL, NK, UA, WN], ignore_index=True)
df
```

### Drop columns with no values

```python
df.dropna(axis=1, how='all')
```

```python
df['ASM'] = df['SEATS']*df['DISTANCE']
df['RPM'] = df['PASSENGERS']*df['DISTANCE']

df['Category'] = "Dom"
df['Dates'] = pd.to_datetime(df.YEAR.astype(str) + '-' + df.MONTH.astype(str))
df
```

## Export the data

```python
import os

pp = "Segment_preprocesseddata2.csv"
path_to_file = os.path.join(root+catchment, pp)
df.to_csv(path_to_file, index=False)

# dataframe struture post conversions
df.info()

print(df.shape)
```

```python
agg_tbl = df.groupby(['Dates','UNIQUE_CARRIER',"Category","YEAR","MONTH"], as_index = False).agg(
     {     
    'DEPARTURES_SCHEDULED':sum,
     'DEPARTURES_PERFORMED':sum,
     'SEATS':sum,
     'PASSENGERS':sum,
     'ASM':sum,
     'RPM':sum     
     }
 )
agg_tbl

agg_tbl=agg_tbl.rename(columns = {'UNIQUE_CARRIER':'Airline','DEPARTURES_SCHEDULED': 'Flights Scheduled','DEPARTURES_PERFORMED': 'Flights','PASSENGERS':'PAX'}) # rename ds column

```

```python
agg_tbl['LF'] = (agg_tbl['RPM']/agg_tbl['ASM'])*100

analysis = "ForecastData2.csv"
path_to_file = os.path.join(root+catchment, analysis)
agg_tbl.to_csv(path_to_file, index=False)
```
