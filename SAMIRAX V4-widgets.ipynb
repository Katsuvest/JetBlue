## Import Packages

```python
# General 
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


# Visualisation
import ipywidgets as widgets # drop down boxes and dashboard
from ipywidgets import interact # as above
import plotly.express as px
```

## Load user defined functions for model

```python
####################
# Widget functions #
####################


# define geneic function to select all values for dropdown boxes (widgets)
ALL = 'ALL'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

SELECT = ''
def unique_sorted_values_plus_SELECT(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, SELECT)
    return unique

# create common filtering process (widgets)
def common_filtering(Airline, Category):
    output.clear_output()
    plot_output.clear_output()
    plot_output_flights.clear_output()
    plot_output_seats.clear_output()
    plot_output_ASM.clear_output()
    plot_output_RPM.clear_output()
    plot_output_LF.clear_output()
    
    if (Airline == SELECT) & (Category == SELECT):  # note:- changed this to select as there is no consoldiated forecast ATM
        common_filter = df
    elif (Airline == SELECT):
        common_filter = df[df.Category == "Dom"]
    elif (Category == SELECT):
        common_filter = df[df.Airline == Airline]
    else:
        common_filter = df[(df.Airline == Airline) & 
                                  (df.Category == Category)]
    with output:
        display(common_filter)
        
    with plot_output:
        fig = px.line(common_filter, x="Datestamp", y="PAX", color="Category")
        fig.update_layout(legend=dict(orientation="h",yanchor="top", y=0.99, xanchor="left", x=0.01))
          
        # Make a vertical highlight section
        fig.add_vrect(x0="2020-03-01", x1="2021-11-08",             
              fillcolor="yellow", opacity=0.25, line_width=0)
        fig.show()  
    
    with plot_output_flights:
        fig = px.line(common_filter, x="Datestamp", y="Flights", color="Category")
        fig.update_layout(legend=dict(orientation="h",yanchor="top", y=0.99, xanchor="left", x=0.01))
        fig.add_vrect(x0="2020-03-01", x1="2021-11-08",             
              fillcolor="yellow", opacity=0.25, line_width=0)
        fig.show()
    with plot_output_seats:
        fig = px.line(common_filter, x="Datestamp", y="SEATS", color="Category")
        fig.update_layout(legend=dict(orientation="h",yanchor="top", y=0.99, xanchor="left", x=0.01))
        fig.add_vrect(x0="2020-03-01", x1="2021-11-08",             
              fillcolor="yellow", opacity=0.25, line_width=0)
        fig.show()          
    with plot_output_ASM:
        fig = px.line(common_filter, x="Datestamp", y="ASM", color="Category")
        fig.update_layout(legend=dict(orientation="h",yanchor="top", y=0.99, xanchor="left", x=0.01))
        fig.add_vrect(x0="2020-03-01", x1="2021-11-08",             
              fillcolor="yellow", opacity=0.25, line_width=0)
        fig.show()   
        
    with plot_output_RPM:
        fig = px.line(common_filter, x="Datestamp", y="RPM", color="Category")
        fig.update_layout(legend=dict(orientation="h",yanchor="top", y=0.99, xanchor="left", x=0.01))
        fig.add_vrect(x0="2020-03-01", x1="2021-11-08",             
              fillcolor="yellow", opacity=0.25, line_width=0)
        fig.show()   
        
    with plot_output_LF:
        fig = px.line(common_filter, x="Datestamp", y="LF", color="Category")
        fig.update_layout(legend=dict(orientation="h",yanchor="top", y=0.99, xanchor="left", x=0.01))
        fig.add_vrect(x0="2020-03-01", x1="2021-11-08",             
              fillcolor="yellow", opacity=0.25, line_width=0)
        fig.show()   
    
# Amend event handlers to call the common filter   (widgets)     
def dropdown_Airline_eventhandler(change):
    common_filtering(change.new, dropdown_Category.value)
    
def dropdown_Category_eventhandler(change):
    common_filtering(dropdown_Airline.value, change.new) 
```

## Import Data

```python
# Import data per selected region

root = r'c:\\users\\user\\Documents\\JCU\\MA 5853 Data Science Project 1\\'
path = r'data\\Airport analysis\\'+Region+'\\'

fileap = r'ForecastData2.csv'   

fileo = r'ForecastDataSelected2.csv'

df = pd.read_csv(root+path+fileap)
df
```

## Wrangle imported data

```python
df['Dates'] = pd.to_datetime(df.YEAR.astype(str) + '-' + df.MONTH.astype(str))

df['Dates'] = pd.DatetimeIndex(df['Dates']) # make index out of ds

df['Datestamp'] = pd.to_datetime(df.YEAR.astype(str) + '-' + df.MONTH.astype(str))

df = df.set_index('Dates') # create index


df.head()
```

## Select Airline and Category for Analysis

```python
# Widgets

# create common output for widgets
output = widgets.Output()
plot_output = widgets.Output()
plot_output_flights = widgets.Output()
plot_output_seats = widgets.Output()
plot_output_ASM = widgets.Output()
plot_output_RPM = widgets.Output()
plot_output_LF = widgets.Output()

# initialise dropdown
dropdown_Airline = widgets.Dropdown(options = unique_sorted_values_plus_SELECT(df.Airline), description='Airline Code:')
dropdown_Category = widgets.Dropdown(options = unique_sorted_values_plus_SELECT(df.Category), description='Category:')

# dropdown boxes actions
dropdown_Airline.observe(dropdown_Airline_eventhandler, names='value')    
dropdown_Category.observe(dropdown_Category_eventhandler, names='value')  

# Layout for dropdown boxes
item_layout = widgets.Layout(margin='0 0 50px 0')

# Alignment and display settings for dropdown boxes
input_widgets = widgets.HBox([dropdown_Airline, dropdown_Category], layout=item_layout)

# dashboard drivers
tab = widgets.Tab([output, plot_output, plot_output_flights, plot_output_seats,  plot_output_ASM, plot_output_RPM, plot_output_LF], layout=item_layout)
tab.set_title(0, 'Dataset Exploration')
tab.set_title(1, 'Passangers')
tab.set_title(2, 'Flights')
tab.set_title(3, 'Seats')
tab.set_title(4, 'ASM')
tab.set_title(5, 'RPM')
tab.set_title(6, 'Load Factor')
```

## Key Performance Indicators Dashboard

```python
dashboard = widgets.VBox([input_widgets, tab])
display(dashboard)
```

## Choose KPI to forecast

### Data Wrangling to prepare selected data for forecasting per KPI

```python
# retrieve the data from the KPI dashboard - note the dashboard is a stage not hard dataframes
df = df[(df['YEAR'] > 2016)]

df2 = df[(df['Airline']==dropdown_Airline.value) &(df['Category']==dropdown_Category.value)] # dataset of kpi display
dfv = columns_select.value # retrieve the value selected to forecast

# create base dataframe to use for forecasting
df1 = df2["Datestamp"]  # datestamp for graphs
df3 = df2[columns_select.value] # select the column of data chosen to forecast
df2 = df2.iloc[:,0:4] # id, dates, airline, category from kpi dataset

df4 = pd.merge(df2,df1, how="inner",on="Dates") # join descriptive columns to datestamp
df4 = pd.merge(df4,df3, how="inner",on="Dates") # join previously joined table to the "value" column


df2 = df4 # rebuild df2 (forecasting dataset)


title2 = df2['Category'].iloc[0]+" "+columns_select.value+" "+df2['Airline'].iloc[0] # create defualt title for graphs
title2 # display the default display
```
